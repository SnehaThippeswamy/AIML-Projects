```python
# Removing outliers and assigning it to new dataframe

Q1 = diabetes_data.quantile(0.25)
Q3 = diabetes_data.quantile(0.75)
IQR = Q3 - Q1

diabetes_data = diabetes_data[~((diabetes_data < (Q1 - 1.5 * IQR)) |(diabetes_data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
diabetes_data.shape
```

#### Do train and test split with stratify sampling on Outcome variable to maintain the distribution of dependent variable(1 Marks)

```python
from sklearn.model_selection import train_test_split
```

```python
X = diabetes_data.drop(columns='Outcome',axis=1)
y = diabetes_data[['Outcome']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, 
                                                    stratify=y, 
                                                    random_state=1)
```

### Train Support Vector Machine Model(1 Marks)

```python
# use from sklearn.svm import SVC
from sklearn import svm
```

```python
svc = svm.SVC(C=.1, kernel='rbf', gamma=1).fit(X_train, y_train)
svc.score(X_test, y_test)
#accuracy scores are low so to get best score we need to fine tune them
```

```python
svc = svm.SVC(C=.1, kernel='linear', gamma=1).fit(X_train, y_train)
svc.score(X_test, y_test)
```

### Scale the data points using MinMaxScaler ( 1 Marks)

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Fit SVM Model on Scale data and give your observation( 1 Marks)

```python
svc = svm.SVC(C=1, kernel='rbf', gamma=1).fit(X_train_scaled, y_train)
svc.score(X_test_scaled, y_test)
#after scaling the data accuracy is increased
```

```python
svc = svm.SVC(C=4, kernel='linear', gamma=1).fit(X_train_scaled, y_train)
svc.score(X_test_scaled, y_test)
```

#we can try increasing either C or gamma to fit a more complex model.

### Try improving the model accuracy using C=1000 (1 Marks)

```python
svc = svm.SVC(C=1000, kernel='rbf', gamma=1).fit(X_train_scaled, y_train)
svc.score(X_test_scaled, y_test)
# Accuracy score is low after increasing
```

### Here, increasing C allows us to improve the model, resulting in lesser accuracy.
