```python
#import packages
import pandas as pd
import numpy as np
```

```python
# Reading Laliga.csv file and use skiprows=1 or header=1 to set column headers with first row
# And storing it in a ddataframe called laliga
laliga = pd.read_csv("Laliga.csv",skiprows=1)
# Using header = 1
# laliga = pd.read_csv("Laliga.csv",header=1)
laliga.head(10)
```

# 1. Replacing '-' with 0

```python
# Replacing - with zero
laliga = laliga.replace('-',0)
laliga.head(10)
```

# 2. All the teams which have started playing between 1930-1980

```python
# Split the column Debut and create new column to store splited data into new columns i.e DebutSplitOfZero and DebutSplitOfOne
laliga['DebutSplitOfZero'], laliga['DebutSplitOfOne'] = laliga['Debut'].str.split('-',1).str
# Apply the condition between 1930 and 1980 on DebutSplitOfZero and Printing the Team Names which are satisfying the condition
laliga[(laliga['DebutSplitOfZero']>='1930') & (laliga['DebutSplitOfZero']<='1980')]['Team']
```

# 3. List of teams which came Top 5 in terms of points 

```python
# Convert Laliga dataframe Points column data into int
laliga['Points'] = pd.to_numeric(laliga['Points'])
# sort the dataframe by points in decending order if its not sorted, then apply loc to retrieve first 5 rows of 'Team' column
laliga.sort_values(by='Points',ascending=False).loc[:4,'Team']
```

#  4. All the teams with their Goal Differences

```python
# Function goal_diff_count which takes 2 parameters one is dataframe another one is requirement as req
# which is set to default req=GoalDiff this will return 'Team' name and 'goalDiffCount', 
# if req="MaxMin" then returns team name of maximum and minimum of goalDiffCount
# if anything else other than "GoalDiff" or "MaxMin" is passed then it'll print Invalid Request
def goal_diff_count(goal_diff,req="GoalDiff"):
    # Converting GoalsFor and GoalsAgainst to numeric datatype
    goal_diff['GoalsFor'] = pd.to_numeric(goal_diff['GoalsFor'])
    goal_diff['GoalsAgainst'] = pd.to_numeric(goal_diff['GoalsAgainst'])
    # Calculating GoalDiffCount, where GoalDiffCount = GoalsFor - GoalsAgainst
    goal_diff['GoalDiffCount'] = goal_diff['GoalsFor']-goal_diff['GoalsAgainst']
    if(req=="GoalDiff"):
        return goal_diff.loc[:,['Team','GoalDiffCount']]
    elif(req=="MaxMin"):
        max = goal_diff[goal_diff['GoalDiffCount'] == goal_diff['GoalDiffCount'].max()]['Team']
        min = goal_diff[goal_diff['GoalDiffCount'] == goal_diff['GoalDiffCount'].min()]['Team']
        return max, min
    else:
        print("Invalid Request!!!!!")
```

```python
#Calling the function by passing laliga Data frame only and requirement is default i.e "GoalDiff" which returns Team with Goal Difference
goal_diff_count(laliga)
```

```python
#Calling the function by passing laliga Data frame and also GoalDiff which returns Team with Goal Difference
goal_diff_count(laliga,"GoalDiff")
```

```python
#Calling the function by passing laliga Data frame and also MaxMin which will return max and min differences
max_diff,min_diff = goal_diff_count(laliga,"MaxMin")
print(f"Max \n\t {max_diff}\nMin \n\t {min_diff}")
```

# 5. Winning Percent

```python
# Converting 'GamesWon' and 'GamesPlayed' to numeric data type
laliga['GamesWon'] = pd.to_numeric(laliga['GamesWon'])
laliga['GamesPlayed'] = pd.to_numeric(laliga['GamesPlayed'])
# Calculating Percentage of wining,where PercantageOfWinning = (GamesWon/GamesPlayed)*100
# And restricting the decimal places to 3 
laliga['PercentageOfWinning(in %)'] = round((laliga['GamesWon']/laliga['GamesPlayed']) * 100,3)
# For 0/0 we get Nan so replacing that Nan with 0 using fillna method
laliga.fillna(0)
# To print Top 5 Teams first we need to sort the data frame by PercentageOfWinning in decending order
# Using loc method printing first 5 rows of 'Team' and 'PercentageOfWinning' Column
laliga.sort_values(by='PercentageOfWinning(in %)',ascending=False).loc[:4,['Team','PercentageOfWinning(in %)']]
```
